## Determine the top 3 most common cuisines in the dataset

```python
##Counting the unique values in cuisines column...

df['Cuisines'].value_counts()
```

```python
##Creating new variable for top 3 cuisines...

Top3_Cuisines = df['Cuisines'].value_counts().head(3)
```

```python
##Visualizing the Data...

plt.figure(figsize=(8,6))
Top3_Cuisines.plot(kind ='bar', color ='yellow', edgecolor ='black')
plt.title("Top 3 Cuisines of the Restaurants")
plt.xlabel('Cuisines')
plt.ylabel("Count")
plt.xticks(rotation=0)  
plt.show()
```

## Calculate the percentage of restaurants that serve each of the top cuisines


```python
##Removing duplicates based on Restaurant Names...

df_unique = df.drop_duplicates(subset=['Restaurant Name'])
```

```python
## Total number of unique restaurants...

total_unique_restaurants = len(df_unique)
total_unique_restaurants
```

```python
# Calculating percentage of unique restaurants serving each of the top 3 cuisines...

percentages = (Top3_Cuisines / total_unique_restaurants) * 100
```

```python
# Display the result
print("\nPercentage of unique restaurants serving each of the top 3 cuisines:")
percentages
```

```python
##Visualizing the Data...

plt.figure(figsize=(8, 6))
plt.bar(percentages.index, percentages.values, color=['blue', 'green', 'red'], edgecolor ='black')
plt.xlabel('Cuisines')
plt.ylabel('Percentage of Restaurants (%)')
plt.title('Percentage of Restaurants Serving Top 3 Cuisines')
plt.ylim(0, 100)
plt.show()
```
