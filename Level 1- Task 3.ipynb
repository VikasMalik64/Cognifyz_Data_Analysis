## Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants

```python
##Display the column labels...

df.columns
```

```python
##Visualizing the Data...

plt.figure(figsize=(8,6))
df['Price range'].value_counts().sort_index().plot(kind='bar', color='yellow', edgecolor='black')
plt.title("DITRIBUTION OF PRICE RANGES AMONG THE RESTAURANTS")
plt.xlabel("Price range")
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

## Calculate the percentage of restaurants in each price range category

```python
# Calculate the counts for each price range...

price_range_counts = df['Price range'].value_counts()
price_range_counts
```

```python
##Visualizing the Data...

plt.figure(figsize=(8,6))

explode = (0.02, 0.02, 0.02, 0.02)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

price_range_counts.plot(kind='pie',labels=price_range_counts.index, autopct='%1.1f%%',explode=explode,colors=colors, startangle=90)
plt.title("PERCENTAGE OF RESTAURANTS IN EACH PRICE PRICE RANGE CATEGORY.")
plt.axis("equal")
plt.show()
```
