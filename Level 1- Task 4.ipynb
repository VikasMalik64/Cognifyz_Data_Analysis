## Determine the percentage of restaurants that offer online delivery

```python
##Display the column labels...

df.columns
```

```python
## Calculate the total number of restaurants...

total_restaurants = len(df)
total_restaurants
```

```python
## Count the number of restaurants that offer online delivery...

restaurants_with_delivery = df['Has Online delivery'].str.lower().eq('yes').sum()
restaurants_with_delivery
```

```python
## Calculate the percentage of restaurants that offer online delivery...

percentage_with_delivery = (restaurants_with_delivery / total_restaurants) * 100
```

```python
##Display Result...

print(f"Percentage of restaurants that offer online delivery: {percentage_with_delivery:.2f}%")
```

```python
##Visualizing the Data...

plt.figure(figsize=(8, 6))
plt.bar(['Restaurants with Online Delivery', 'Restaurants without Online Delivery'], [percentage_with_delivery, 100 - percentage_with_delivery], color ='yellow', edgecolor ='black')
plt.title('Percentage of Restaurants Offering Online Delivery')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.show()
```

## Compare the average ratings of restaurants with and without onine delivery

```python
##Display the column labels...

df.columns
```

```python
##Group the data by delivery and calculate the average rating...

restraurant_avg_rating = df.groupby('Has Online delivery')["Aggregate rating"].mean()
```

```python
##Display Result...

print("Avg Rating of Restaurants with and without Online Delivery:")
print(restraurant_avg_rating)
```

```python
##Visualizing the Data...

restraurant_avg_rating.plot(kind='bar', color = '#1f77b4', edgecolor ='black')
plt.title("Average Ratings of Restaurants")
plt.xlabel("Restaurant With Online ")
plt.ylabel("Average Rating")
plt.xticks([0,1], ['No','Yes'] ,rotation=0)
plt.show()
```
