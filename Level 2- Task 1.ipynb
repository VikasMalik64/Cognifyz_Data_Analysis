## Analyze the distribution of agregate ratings and determine the most common rating range

```python
##Display the column labels...

df.columns
```

```python
# Determine the most common rating range

rating_counts = df['Aggregate rating'].value_counts()

most_common_rating = rating_counts.idxmax()
most_common_rating_count = rating_counts.max()
```

```python
##Visualizing the Data...

plt.hist(df['Aggregate rating'], bins = 5,color ='yellow', edgecolor='black')
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Aggregate Ratings")
plt.xticks(rotation=0)
plt.show()
```
